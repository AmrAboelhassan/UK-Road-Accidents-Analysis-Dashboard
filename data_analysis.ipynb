```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile
import os

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
zip_name = list(uploaded.keys())[0]

with zipfile.ZipFile(zip_name, 'r') as zip_ref:
    zip_ref.extractall("road_safety_data")

print("Files extracted successfully!")
print(os.listdir("road_safety_data"))
```

```python
accidents = pd.read_csv(
    "road_safety_data/Accident_Information.csv",
    low_memory=False,
    encoding="latin1"
)

vehicles = pd.read_csv(
    "road_safety_data/Vehicle_Information.csv",
    low_memory=False,
    encoding="latin1"
)

print("Accidents shape:", accidents.shape)
print("Vehicles shape:", vehicles.shape)
```

```python
accidents_clean = accidents.copy()
vehicles_clean = vehicles.copy()

print("Copies created successfully")
```

```python
accidents_clean["Date"] = pd.to_datetime(
    accidents_clean["Date"],
    errors="coerce"
)

accidents_clean["Time"] = pd.to_datetime(
    accidents_clean["Time"],
    format="%H:%M",
    errors="coerce"
)

accidents_clean["Year"] = accidents_clean["Date"].dt.year
accidents_clean["Month"] = accidents_clean["Date"].dt.month
accidents_clean["Day_Name"] = accidents_clean["Date"].dt.day_name()
accidents_clean["Hour"] = accidents_clean["Time"].dt.hour

accidents_clean[["Date", "Time", "Year", "Month", "Day_Name", "Hour"]].head()
```

```python
severity_map = {
    1: "Fatal",
    2: "Serious",
    3: "Slight",
    "1": "Fatal",
    "2": "Serious",
    "3": "Slight",
    "Fatal": "Fatal",
    "Serious": "Serious",
    "Slight": "Slight"
}

accidents_clean["Severity_Label"] = accidents_clean["Accident_Severity"].map(severity_map)

accidents_clean["Severity_Label"] = accidents_clean["Severity_Label"].fillna(
    accidents_clean["Accident_Severity"].astype(str)
)

accidents_clean["Severity_Label"].value_counts()
```

```python
acc_missing = accidents_clean.isnull().mean().sort_values(ascending=False) * 100
veh_missing = vehicles_clean.isnull().mean().sort_values(ascending=False) * 100

print("Top missing values in accidents:")
print(acc_missing[acc_missing > 0].head(15))

print("\nTop missing values in vehicles:")
print(veh_missing[veh_missing > 0].head(15))
```

```python
top_acc_missing = acc_missing[acc_missing > 0].head(15)

plt.figure(figsize=(12,6))
sns.barplot(x=top_acc_missing.values, y=top_acc_missing.index)

plt.title("Top Missing Values in Accidents Dataset")
plt.xlabel("Missing Percentage (%)")
plt.ylabel("Columns")
plt.show()
```

```python
top_veh_missing = veh_missing[veh_missing > 0].head(15)

plt.figure(figsize=(12,6))
sns.barplot(x=top_veh_missing.values, y=top_veh_missing.index)

plt.title("Top Missing Values in Vehicles Dataset")
plt.xlabel("Missing Percentage (%)")
plt.ylabel("Columns")
plt.show()
```

```python
yearly_accidents = accidents_clean.groupby("Year").size()

plt.figure(figsize=(12,6))
sns.lineplot(x=yearly_accidents.index, y=yearly_accidents.values, marker="o")

plt.title("Number of Road Accidents by Year")
plt.xlabel("Year")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
severity_order = ["Slight", "Serious", "Fatal"]

plt.figure(figsize=(8,6))
sns.countplot(
    data=accidents_clean,
    x="Severity_Label",
    order=[x for x in severity_order if x in accidents_clean["Severity_Label"].unique()]
)

plt.title("Accident Severity Distribution")
plt.xlabel("Accident Severity")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
monthly_accidents = accidents_clean.groupby("Month").size()

plt.figure(figsize=(10,6))
sns.barplot(x=monthly_accidents.index, y=monthly_accidents.values)

plt.title("Number of Accidents by Month")
plt.xlabel("Month")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
day_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

plt.figure(figsize=(12,6))
sns.countplot(
    data=accidents_clean,
    x="Day_Name",
    order=day_order
)

plt.title("Number of Accidents by Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.show()
```

```python
hourly_accidents = accidents_clean.groupby("Hour").size()

plt.figure(figsize=(12,6))
sns.lineplot(x=hourly_accidents.index, y=hourly_accidents.values, marker="o")

plt.title("Number of Accidents by Hour of Day")
plt.xlabel("Hour")
plt.ylabel("Number of Accidents")
plt.xticks(range(0, 24))
plt.show()
```

```python
heatmap_data = accidents_clean.pivot_table(
    index="Day_Name",
    columns="Hour",
    values="Accident_Index",
    aggfunc="count"
)

heatmap_data = heatmap_data.reindex(day_order)

plt.figure(figsize=(15,6))
sns.heatmap(heatmap_data, cmap="YlOrRd")

plt.title("Accidents Heatmap by Day of Week and Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Day of Week")
plt.show()
```

```python
weather_counts = accidents_clean["Weather_Conditions"].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=weather_counts.values, y=weather_counts.index)

plt.title("Top Weather Conditions During Accidents")
plt.xlabel("Number of Accidents")
plt.ylabel("Weather Conditions")
plt.show()
```

```python
weather_severity = pd.crosstab(
    accidents_clean["Weather_Conditions"],
    accidents_clean["Severity_Label"]
)

weather_severity = weather_severity.loc[weather_severity.sum(axis=1).sort_values(ascending=False).head(8).index]

weather_severity.plot(kind="bar", stacked=True, figsize=(14,7))

plt.title("Accident Severity by Weather Conditions")
plt.xlabel("Weather Conditions")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
road_surface_counts = accidents_clean["Road_Surface_Conditions"].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=road_surface_counts.values, y=road_surface_counts.index)

plt.title("Accidents by Road Surface Conditions")
plt.xlabel("Number of Accidents")
plt.ylabel("Road Surface Conditions")
plt.show()
```

```python
light_counts = accidents_clean["Light_Conditions"].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x=light_counts.values, y=light_counts.index)

plt.title("Accidents by Light Conditions")
plt.xlabel("Number of Accidents")
plt.ylabel("Light Conditions")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(
    data=accidents_clean,
    x="Speed_limit",
    hue="Severity_Label"
)

plt.title("Accident Severity by Speed Limit")
plt.xlabel("Speed Limit")
plt.ylabel("Number of Accidents")
plt.legend(title="Severity")
plt.show()
```

```python
urban_counts = accidents_clean["Urban_or_Rural_Area"].value_counts()

plt.figure(figsize=(8,6))
sns.barplot(x=urban_counts.index, y=urban_counts.values)

plt.title("Urban vs Rural Accidents")
plt.xlabel("Area Type")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.show()
```

```python
casualties_by_year = accidents_clean.groupby("Year")["Number_of_Casualties"].sum()

plt.figure(figsize=(12,6))
sns.lineplot(x=casualties_by_year.index, y=casualties_by_year.values, marker="o")

plt.title("Total Casualties by Year")
plt.xlabel("Year")
plt.ylabel("Number of Casualties")
plt.show()
```

```python
avg_casualties = accidents_clean.groupby("Severity_Label")["Number_of_Casualties"].mean().sort_values(ascending=False)

plt.figure(figsize=(8,6))
sns.barplot(x=avg_casualties.index, y=avg_casualties.values)

plt.title("Average Number of Casualties per Accident by Severity")
plt.xlabel("Accident Severity")
plt.ylabel("Average Casualties")
plt.show()
```

```python
vehicles_involved = accidents_clean["Number_of_Vehicles"].value_counts().sort_index().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=vehicles_involved.index, y=vehicles_involved.values)

plt.title("Number of Vehicles Involved in Accidents")
plt.xlabel("Number of Vehicles")
plt.ylabel("Number of Accidents")
plt.show()
```

```python
driver_sex_counts = vehicles_clean["Sex_of_Driver"].value_counts()

plt.figure(figsize=(8,6))
sns.barplot(x=driver_sex_counts.index, y=driver_sex_counts.values)

plt.title("Accidents by Sex of Driver")
plt.xlabel("Sex of Driver")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=45)
plt.show()
```

```python
age_band_counts = vehicles_clean["Age_Band_of_Driver"].value_counts().head(15)

plt.figure(figsize=(12,6))
sns.barplot(x=age_band_counts.values, y=age_band_counts.index)

plt.title("Driver Age Band Involved in Accidents")
plt.xlabel("Number of Vehicles")
plt.ylabel("Driver Age Band")
plt.show()
```

```python
manoeuvre_counts = vehicles_clean["Vehicle_Manoeuvre"].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=manoeuvre_counts.values, y=manoeuvre_counts.index)

plt.title("Top Vehicle Manoeuvres During Accidents")
plt.xlabel("Number of Vehicles")
plt.ylabel("Vehicle Manoeuvre")
plt.show()
```

```python
merged_data = vehicles_clean.merge(
    accidents_clean[[
        "Accident_Index",
        "Severity_Label",
        "Weather_Conditions",
        "Road_Surface_Conditions",
        "Light_Conditions",
        "Speed_limit",
        "Urban_or_Rural_Area",
        "Year"
    ]],
    on="Accident_Index",
    how="left"
)

print("Merged data shape:", merged_data.shape)
merged_data.head()
```

```python
top_types = merged_data["Vehicle_Type"].value_counts().head(8).index

vehicle_severity = pd.crosstab(
    merged_data[merged_data["Vehicle_Type"].isin(top_types)]["Vehicle_Type"],
    merged_data[merged_data["Vehicle_Type"].isin(top_types)]["Severity_Label"]
)

vehicle_severity.plot(kind="bar", stacked=True, figsize=(14,7))

plt.title("Accident Severity by Vehicle Type")
plt.xlabel("Vehicle Type")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
top_age_bands = merged_data["Age_Band_of_Driver"].value_counts().head(10).index

age_severity = pd.crosstab(
    merged_data[merged_data["Age_Band_of_Driver"].isin(top_age_bands)]["Age_Band_of_Driver"],
    merged_data[merged_data["Age_Band_of_Driver"].isin(top_age_bands)]["Severity_Label"]
)

age_severity.plot(kind="bar", stacked=True, figsize=(14,7))

plt.title("Accident Severity by Driver Age Band")
plt.xlabel("Driver Age Band")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
sex_severity = pd.crosstab(
    merged_data["Sex_of_Driver"],
    merged_data["Severity_Label"]
)

sex_severity.plot(kind="bar", stacked=True, figsize=(10,6))

plt.title("Accident Severity by Sex of Driver")
plt.xlabel("Sex of Driver")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation=45)
plt.show()
```
